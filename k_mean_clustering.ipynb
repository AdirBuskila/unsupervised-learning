```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bank-full.csv')
df.head()
```

```python
plt.figure(figsize=(12, 6), dpi=200)
sns.histplot(data=df, x="age", bins=40, hue="loan")
plt.show()
```

```python
data_df = df[df['pdays'] != 999]
plt.figure(figsize=(12, 6), dpi=200)
sns.histplot(data=data_df, x="pdays", bins=40)
plt.show()
```

```python
plt.figure(figsize=(12, 6), dpi=200)
sns.histplot(data=data_df, x="duration", bins=100, hue="contact")
plt.show()
```

```python
order_edu_df = df["job"].value_counts().index

plt.figure(figsize=(12, 6), dpi=200)
sns.countplot(data=df, x="job", order=order_edu_df,hue='job')
plt.xticks(rotation=90)
plt.show()
```

```python
order_edu_df = df["education"].value_counts().index

plt.figure(figsize=(12, 6), dpi=200)
sns.countplot(data=df, x="education", order=order_edu_df, hue="education")
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X = pd.get_dummies(df)
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)

model = KMeans(n_clusters=2)

cluster_labels = model.fit_predict(scaled_X)
cluster_labels
```
