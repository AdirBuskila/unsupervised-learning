```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("wine_dataset.csv")
df.head()
```

```python
plt.figure(figsize=(12, 6), dpi=120)
sns.histplot(data=df, x="alcohol", bins=30)
plt.show()
```

```python
plt.figure(figsize=(12, 6), dpi=120)
sns.histplot(data=df, x="proline", bins=30)
plt.show()
```

```python
df.hist(bins=30, figsize=(14, 10))
plt.show()
```

```python
plt.figure(figsize=(10, 8), dpi=120)
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=120)
sns.scatterplot(data=df, x="flavanoids", y="color_intensity")
plt.show()
```

```python
sns.pairplot(df[["alcohol", "flavanoids", "color_intensity", "proline"]])
plt.show()
```

```python
plt.figure(figsize=(14, 6), dpi=120)
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```
